# Bài 1: Thực hiện lại các câu hỏi (Câu 1 đến Câu 5) trong mục 2 để xây dựng lại mô hình hồi quy tuyến tính.

```python
# 1

import matplotlib.pyplot as plt 
import numpy as np 

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

X_b = np.c_[np.ones((100, 1)), X]

plt.scatter(X, y)
plt.show()
```

```python
# 2

print(X_b.shape)
print(y.shape)
```

```python
# 3

# Tính theta tối ưu
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# In ra theta tối ưu
print("Theta tối ưu:", theta_best)
```

```python
# 4

# Tạo dữ liệu mới để kiểm tra
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]

# Dự đoán giá trị dự báo cho dữ liệu mới
y_pred = X_new_b.dot(theta_best)
print("Dự đoán:", y_pred)
```

```python
# 5

import matplotlib.pyplot as plt

# Vẽ biểu đồ cho dữ liệu huấn luyện ban đầu
plt.plot(X, y, "b.", label='Dữ liệu huấn luyện')

# Vẽ biểu đồ cho dự đoán của mô hình hồi quy tuyến tính
plt.plot(X_new, y_pred, "r-", label='Dự đoán')

plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

# Bài 2: Thực hiện xây dựng mô hình hồi quy đa thức (Polynomial Regression) cho dữ liệu đa thức như sau:


```python
# Dữ liệu huấn luyện:
m=100
X_2 = 6 * np.random.rand(m, 1) - 3
y_2 = 0.5 * X_2**2 + X_2 + 2 + np.random.randn(m, 1)

# Dữ liệu huấn luyện
X_new = np.linspace(-3, 3, 100).reshape(100, 1)

plt.scatter(X_2, y_2)
plt.show()

from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2,
include_bias=False)

X_poly = poly_features.fit_transform(X_2)
X_new_poly = poly_features.fit_transform(X_new)

```

```python
print(X_poly.shape)
print(y_2.shape)
theta_best_2 = np.linalg.inv(X_poly.T.dot(X_poly)).dot(X_poly.T).dot(y_2)

# In ra theta tối ưu
print("Theta tối ưu:", theta_best_2)
```

```python
# Dự đoán giá trị dự báo cho dữ liệu mới
y_pred_2 = X_new_poly.dot(theta_best_2)
print("Dự đoán:", y_pred_2)
```

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ cho dữ liệu huấn luyện ban đầu
plt.plot(X_2, y_2, "b.", label='Dữ liệu huấn luyện')

# Vẽ biểu đồ cho dự đoán của mô hình hồi quy tuyến tính
plt.plot(X_new, y_pred, "r-", label='Dự đoán')

plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

# Bài 3: Hãy xây dựng mô hình hồi quy Ridge. Thực hiện tương tự như Bài 1.


```python

import matplotlib.pyplot as plt 
import numpy as np 

X_3 = 2 * np.random.rand(100, 1)
y_3 = 4 + 3 * X_3 + np.random.randn(100, 1)

X_b_3 = np.c_[np.ones((100, 1)), X_3]

plt.scatter(X_3, y_3)
plt.show()

alpha = 0.02
I = np.identity(2)  # Đây là 

# Tính theta tối ưu với alpha regularization
theta_best_3 = np.dot(np.dot(np.linalg.inv(np.dot(X_b_3.T, X_b_3)+alpha*I), X_b_3.T), y_3)

# In ra theta tối ưu
print("Theta tối ưu với alpha regularization:", theta_best_3)

# Tạo dữ liệu mới để kiểm tra
X_new_3 = np.array([[0], [2]])
X_new_b_3 = np.c_[np.ones((2, 1)), X_new_3]

# Dự đoán giá trị dự báo cho dữ liệu mới
y_pred_3 = X_new_b_3.dot(theta_best_3)
print("Dự đoán:", y_pred_3)
```

```python
plt.plot(X_3, y_3,'b.')
plt.plot(X_new_3, y_pred_3, 'r-')
```

# Bài 4: Các bạn hãy so sánh giữa tham số tối ưu tìm được bằng tay và bằng thư viện sklearn trong Bài 1 và Bài 3. So sánh kết quả thu được giữa thực hiện bằng tay và bằng thư viện.

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(X_b, y)

# Lấy hệ số góc (slope)
slope = reg.coef_

print("Hệ số góc (slope):", slope)
```

# Nhận xét: 
Dựa trên kết quả bạn đã cung cấp, có vẻ như có sự khác biệt đáng kể giữa các tham số tối ưu được tìm thấy bằng cách sử dụng hàm tự viết và hàm của thư viện sklearn. 
Cụ thể:

Tham số tối ưu bằng hàm tự viết:
[4.00895978 3.03681045​]

Tham số tối ưu bằng hàm sklearn:
[0.0 2.52895801​]

Bài 3:
Tham số tối ưu bằng hàm tự viết:
[4.27589163 2.66064094]

Tham số tối ưu bằng hàm sklearn:
[0.0 2.58200124​]

Có một số nguyên nhân có thể dẫn đến sự khác biệt này:
Dữ liệu: Sklearn thường có các bước tiền xử lý dữ liệu như chuẩn hóa. Nếu bạn không thực hiện các bước tương tự khi thực hiện bằng tay, kết quả có thể khác nhau.
Điều kiện dừng: Sklearn sử dụng một số tiêu chí dừng nhất định cho các thuật toán tối ưu hóa của nó. Nếu bạn sử dụng một tiêu chí dừng khác nhau, thuật toán của bạn có thể chạy lâu hơn hoặc ngắn hơn và do đó đạt được một kết quả khác nhau.
Tính chính xác số học: Do sai số làm tròn và tính chất của các thuật toán số học, việc thực hiện cùng một thuật toán trên hai hệ thống khác nhau có thể dẫn đến các kết quả khác nhau.
